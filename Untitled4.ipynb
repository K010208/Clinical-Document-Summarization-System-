```python
!pip install -q transformers torch pypdf rouge-score

import torch
from transformers import pipeline
from rouge_score import rouge_scorer
from pypdf import PdfReader

clinical_text = """
Pembrolizumab is a monoclonal antibody used in cancer immunotherapy.
Clinical trials indicate improved survival in melanoma patients.
Adverse effects include fatigue, nausea, diarrhea, and immune-mediated reactions.
Careful monitoring is required due to the risk of pneumonitis and colitis.
"""


def clean_text(text, max_sentences=40):
    sentences = text.replace("\n", " ").split(". ")
    return ". ".join(sentences[:max_sentences])

cleaned_text = clean_text(clinical_text)


summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn",
    device=0 if torch.cuda.is_available() else -1
)


summary = summarizer(
    cleaned_text,
    max_length=150,
    min_length=60,
    do_sample=False
)[0]["summary_text"]

print("ðŸ§  SUMMARY:\n")
print(summary)


def relevance_check(original, summary, threshold=0.25):
    original_words = set(original.lower().split())
    summary_words = set(summary.lower().split())
    overlap = len(original_words & summary_words) / max(len(summary_words), 1)
    return overlap >= threshold, overlap

is_relevant, relevance_score = relevance_check(cleaned_text, summary)

print("\nâœ… Relevant Summary:", is_relevant)
print("ðŸ” Relevance Score:", round(relevance_score, 2))

compression_ratio = 1 - (len(summary.split()) / len(cleaned_text.split()))

print("\nðŸ“‰ Compression:", round(compression_ratio * 100, 2), "%")

scorer = rouge_scorer.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)
scores = scorer.score(cleaned_text, summary)

print("\nðŸ“Š ROUGE SCORES")
print("ROUGE-1:", round(scores["rouge1"].fmeasure, 3))
print("ROUGE-L:", round(scores["rougeL"].fmeasure, 3))
```

```python
import streamlit as st
import torch
from transformers import pipeline
from pypdf import PdfReader
from rouge_score import rouge_scorer

def read_pdf(file):
    reader = PdfReader(file)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text


def clean_text(text, max_sentences=40):
    sentences = text.replace("\n", " ").split(". ")
    return ". ".join(sentences[:max_sentences])


def relevance_check(original, summary, threshold=0.25):
    original_words = set(original.lower().split())
    summary_words = set(summary.lower().split())
    overlap = len(original_words & summary_words) / max(len(summary_words), 1)
    return overlap >= threshold, overlap



@st.cache_resource
def load_model():
    return pipeline(
        "summarization",
        model="facebook/bart-large-cnn",
        device=0 if torch.cuda.is_available() else -1
    )

summarizer = load_model()


st.set_page_config(page_title="Clinical Document Summarizer", layout="centered")

st.title("ðŸ¥ Clinical Document Summarization System")
st.write("Summarize clinical reports and research documents with reduced hallucination.")

uploaded_file = st.file_uploader("Upload Clinical PDF", type=["pdf"])

if uploaded_file:
    with st.spinner("Reading document..."):
        raw_text = read_pdf(uploaded_file)
        cleaned_text = clean_text(raw_text)

    with st.spinner("Generating summary..."):
        summary = summarizer(
            cleaned_text,
            max_length=150,
            min_length=60,
            do_sample=False
        )[0]["summary_text"]

    is_relevant, relevance_score = relevance_check(cleaned_text, summary)

    compression = 1 - (len(summary.split()) / len(cleaned_text.split()))

    scorer = rouge_scorer.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)
    scores = scorer.score(cleaned_text, summary)

    st.subheader("ðŸ§  Summary")
    st.write(summary)

    st.subheader("ðŸ“Š Quality Metrics")
    st.metric("Compression (%)", f"{round(compression*100,2)}")
    st.metric("Relevance Score", round(relevance_score, 2))
    st.metric("ROUGE-1", round(scores["rouge1"].fmeasure, 3))
    st.metric("ROUGE-L", round(scores["rougeL"].fmeasure, 3))
```

```python
%%writefile app.py
import streamlit as st
import torch
from transformers import pipeline
import nltk
from nltk.tokenize import sent_tokenize
from PyPDF2 import PdfReader


nltk.download('punkt')

st.set_page_config(page_title="Clinical Document Summarizer", layout="wide")
st.title("ðŸ©º Clinical Document Summarization System")

@st.cache_resource
def load_model():
    return pipeline(
        "summarization",
        model="facebook/bart-large-cnn",
        device=0 if torch.cuda.is_available() else -1
    )

summarizer = load_model()

def clean_text(text, max_sentences=40):
    sentences = sent_tokenize(text)
    return " ".join(sentences[:max_sentences])

uploaded_file = st.file_uploader("Upload Clinical PDF", type=["pdf"])

if uploaded_file:
    reader = PdfReader(uploaded_file)
    clinical_text = " ".join([page.extract_text() for page in reader.pages])

    cleaned_text = clean_text(clinical_text)

    if st.button("Generate Summary"):
        with st.spinner("Summarizing..."):
            summary = summarizer(
                cleaned_text,
                max_length=180,
                min_length=80,
                do_sample=False
            )[0]["summary_text"]

        st.subheader("ðŸ“„ Summary")
        st.write(summary)
```
